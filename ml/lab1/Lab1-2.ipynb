6. Как вычислить ((A+B)*(-A/2)) на месте (без копирования)?

````python
A = np.array([1, 2], dtype=float)
B = np.array([3, 4], dtype=float)
np.add(A, B, out=B)      
np.negative(A, out=A)   
np.divide(A, 2, out=A)  
np.multiply(B, A, out=B) 
````

### 3. Визуализация

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

1. Постройте диаграмму которая отражает средний уровень лояльности во всех регионах

````python
import matplotlib.pyplot as plt
````

````python
mean_loyalty_by_region = customer_data.groupby('region')['loyalty_score'].mean()

plt.figure(figsize=(10, 6))
mean_loyalty_by_region.plot(kind='bar', color='skyblue')
plt.title('Средний уровень лояльности по регионам')
plt.xlabel('Регион')
plt.ylabel('Средний уровень лояльности')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.tight_layout()
plt.show()
````

2. Постройте гистограмму распределения зарплат клиентов на Севере.

````python
customer_data.query('region == "North"')['annual_income']\
    .hist(bins = int(np.log(len(customer_data.query('region == "North"')['annual_income'].unique()))) + 1)
plt.title('Распределения зарплат клиентов на Севере')
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.show()
````

3. Сравните уровни зарплат людей на севере и юге.

````python
north_income = customer_data.query('region == "North"')['annual_income']
south_income = customer_data.query('region == "South"')['annual_income']

plt.figure(figsize=(12, 6))

plt.hist(north_income, bins=int(np.log(len(north_income.unique()))) + 1, alpha=0.5, label='Север', color='blue')

plt.hist(south_income, bins=int(np.log(len(south_income.unique()))) + 1, alpha=0.5, label='Юг', color='orange')

plt.title('Сравнение уровней зарплат клиентов на Севере и Юге')
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.legend()

plt.show()
````

4. Постройте box plot отображающий распределение уровня лояльности для людей до 20 лет, от 20 до 40, от 40 до 60 и старше 60 лет

````python
diap = [(0, 20), (20, 40), (40, 60), (60, np.inf)]

for lower, upper in diap:
    filtered_data = customer_data[
        (customer_data['age'] >= lower) & 
        (customer_data['age'] < upper)
    ]
    plt.figure(figsize=(6, 4))
    filtered_data['loyalty_score'].plot.box()
    plt.title(f'Лояльность для возраста {lower}-{upper}')
    plt.ylabel('Оценка лояльности')
    plt.show()
````

5. Визуализируйте зависимость частоты покупок от заработка
